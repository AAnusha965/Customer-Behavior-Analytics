```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.columns=df.columns.str.replace('_','')
```

```python
df=df.rename(columns={'purchaseamount(usd)':'purchaseamount'})
```

```python
#create a columns age group
labels = ['young_adult','adult','middle-aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create columns purchase frequency days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequencyofpurchases'].map(frequency_mapping)
```
